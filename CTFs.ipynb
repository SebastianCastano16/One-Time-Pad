```python
#oncetimepad

import base64
import random
import time

def generate_one_time_pad(seed, length):
    random.seed(seed)
    return bytes([random.randint(0, 255) for _ in range(length)])

def decrypt_message(encrypted_base64, one_time_pad):
    encrypted_message = base64.b64decode(encrypted_base64)
    return bytes([c ^ k for c, k in zip(encrypted_message, one_time_pad)])

# Timestamp without seconds
# Example: "Tue, 28 Nov 2023 23:26"
timestamp_str = "Fri, 01 Dec 2023 00:33"

# Encrypted message (base64 encoded)
encrypted_message_base64 = "AduTUm/qRfWHJJSrJR+z3aqhw8hrAfPntlofJw6flR7LCM2MR1Mhj6IqshCbKp3foKn9NYo="
# Trying all 60 possible seconds in the given minute
for sec in range(60):
    full_timestamp_str = f"{timestamp_str}:{sec}"
    timestamp = time.mktime(time.strptime(full_timestamp_str, "%a, %d %b %Y %H:%M:%S"))

    # Generate the one-time pad
    one_time_pad = generate_one_time_pad(int(timestamp), len(base64.b64decode(encrypted_message_base64)))

    # Decrypt the message
    decrypted_message = decrypt_message(encrypted_message_base64, one_time_pad)

    print(f"Second: {sec}, Decrypted Message: {decrypted_message.decode(errors='ignore')}")
```
